# Plot results for paper

```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import dataset of query times
df = pd.read_csv('query_times_finals.csv')
df.head()
```

```python
#log transform because of extreme value difference
df['log_time'] = np.log(df['time'])
df.head()
```

```python
# who v/s fare barplot
plt.rcParams["figure.figsize"] = [15, 10]
sns.barplot(x = 'query_name',
            y = 'log_time',
            data = df)
plt.xticks(rotation=45, fontsize = 12) 
plt.yticks(fontsize = 12) 
plt. xlabel('TPC-H Query', fontsize = 14)
plt. ylabel('Log Time', fontsize = 14)
# Show the plot
plt.show()
```

```python
#plot comparison of optimized query with original
comp = pd.DataFrame(columns=['Query Name', 'Run Time'])
comp = comp.append({'Query Name': 'Original Query 20', 'Run Time': df.loc[df.query_name == 'Query 20', 'time'].item()}, ignore_index=True)
comp = comp.append({'Query Name': 'Optimized Query 20', 'Run Time': 4.444279}, ignore_index=True)
comp
#df.loc[df.query_name == 'Query 20', 'time'].item()
```

```python
comp['log_time'] = np.log(comp['Run Time'])
comp.head()
```

```python
#plot original run times
plt.rcParams["figure.figsize"] = [5, 5]
sns.barplot(x = 'Query Name',
            y = 'Run Time',
            data = comp,
           dodge=False)
plt.xticks(fontsize = 12) 
plt.yticks(fontsize = 12) 
plt. xlabel(' ')
plt. ylabel('Run Time (seconds)', fontsize = 14)
# Show the plot
plt.show()
```

```python
#plot log of run times
plt.rcParams["figure.figsize"] = [5, 5]
sns.barplot(x = 'Query Name',
            y = 'log_time',
            data = comp,
           dodge=False)
plt.xticks(fontsize = 12) 
plt.yticks(fontsize = 12) 
plt. xlabel(' ')
plt. ylabel('Run Time (seconds)', fontsize = 14)
# Show the plot
plt.show()
```
